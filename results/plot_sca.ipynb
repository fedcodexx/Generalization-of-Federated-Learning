```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import StrMethodFormatter
```

```python
scaffold_test_iid = pd.read_csv('iid/mnist_SCAFFOLD_cnn_00.003_test.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_test_iid = np.array(scaffold_test_iid['scaffold'].to_list())

scaffold_train_iid = pd.read_csv('iid/mnist_SCAFFOLD_cnn_00.003_train.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_train_iid = np.array(scaffold_train_iid['scaffold'].to_list())

scaffold_test_pat = pd.read_csv('pat/mnist_SCAFFOLD_cnn_00.003_test.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_test_pat = np.array(scaffold_test_pat['scaffold'].to_list())

scaffold_train_pat = pd.read_csv('pat/mnist_SCAFFOLD_cnn_00.003_train.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_train_pat = np.array(scaffold_train_pat['scaffold'].to_list())

scaffold_test_08 = pd.read_csv('pat_0.8/mnist_SCAFFOLD_cnn_00.003_test.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_test_08 = np.array(scaffold_test_08['scaffold'].to_list())

scaffold_train_08 = pd.read_csv('pat_0.8/mnist_SCAFFOLD_cnn_00.003_train.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_train_08 = np.array(scaffold_train_08['scaffold'].to_list())

scaffold_test_05 = pd.read_csv('pat_0.5/mnist_SCAFFOLD_cnn_00.003_test.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_test_05 = np.array(scaffold_test_05['scaffold'].to_list())

scaffold_train_05 = pd.read_csv('pat_0.5/mnist_SCAFFOLD_cnn_00.003_train.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_train_05 = np.array(scaffold_train_05['scaffold'].to_list())

scaffold_test_02 = pd.read_csv('pat_0.2/mnist_SCAFFOLD_cnn_00.003_test.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_test_02 = np.array(scaffold_test_02['scaffold'].to_list())

scaffold_train_02 = pd.read_csv('pat_0.2/mnist_SCAFFOLD_cnn_00.003_train.csv', usecols=[1], header = 0, names=['scaffold'])
scaffold_train_02 = np.array(scaffold_train_02['scaffold'].to_list())
```

```python
ls1 = [0, 0.2, 0.5, 0.8, 1]
```

```python
avg_02 = []
avg_008 = []
avg_005 = []
avg_001 = []
avg_0005 = []

c = 0.2
for i in range(len(scaffold_train_iid)):
    if scaffold_train_iid[i] < c:
        avg_02.append(abs(scaffold_test_iid[i]-scaffold_train_iid[i]))
        break

for i in range(len(scaffold_train_02)):
    if scaffold_train_02[i] < c:
        avg_02.append(abs(scaffold_test_02[i]-scaffold_train_02[i]))
        break

for i in range(len(scaffold_train_05)):
    if scaffold_train_05[i] < c:
        avg_02.append(abs(scaffold_test_05[i]-scaffold_train_05[i]))
        break

for i in range(len(scaffold_train_08)):
    if scaffold_train_08[i] < c:
        avg_02.append(abs(scaffold_test_08[i]-scaffold_train_08[i]))
        break
for i in range(len(scaffold_train_pat)):
    if scaffold_train_pat[i] < c:
        avg_02.append(abs(scaffold_test_pat[i]-scaffold_train_pat[i]))
        break

c = 0.1
for i in range(len(scaffold_train_iid)):
    if scaffold_train_iid[i] < c:
        avg_008.append(abs(scaffold_test_iid[i]-scaffold_train_iid[i]))
        break

for i in range(len(scaffold_train_02)):
    if scaffold_train_02[i] < c:
        avg_008.append(abs(scaffold_test_02[i]-scaffold_train_02[i]))
        break

for i in range(len(scaffold_train_05)):
    if scaffold_train_05[i] < c:
        avg_008.append(abs(scaffold_test_05[i]-scaffold_train_05[i]))
        break

for i in range(len(scaffold_train_08)):
    if scaffold_train_08[i] < c:
        avg_008.append(abs(scaffold_test_08[i]-scaffold_train_08[i]))
        break
for i in range(len(scaffold_train_pat)):
    if scaffold_train_pat[i] < c:
        avg_008.append(abs(scaffold_test_pat[i]-scaffold_train_pat[i]))
        break

c = 0.05
for i in range(len(scaffold_train_iid)):
    if scaffold_train_iid[i] < c:
        avg_005.append(abs(scaffold_test_iid[i]-scaffold_train_iid[i]))
        break

for i in range(len(scaffold_train_02)):
    if scaffold_train_02[i] < c:
        avg_005.append(abs(scaffold_test_02[i]-scaffold_train_02[i]))
        break

for i in range(len(scaffold_train_05)):
    if scaffold_train_05[i] < c:
        avg_005.append(abs(scaffold_test_05[i]-scaffold_train_05[i]))
        break

for i in range(len(scaffold_train_08)):
    if scaffold_train_08[i] < c:
        avg_005.append(abs(scaffold_test_08[i]-scaffold_train_08[i]))
        break
for i in range(len(scaffold_train_pat)):
    if scaffold_train_pat[i] < c:
        avg_005.append(abs(scaffold_test_pat[i]-scaffold_train_pat[i]))
        break

c = 0.01
for i in range(len(scaffold_train_iid)):
    if scaffold_train_iid[i] < c:
        avg_001.append(abs(scaffold_test_iid[i]-scaffold_train_iid[i]))
        break

for i in range(len(scaffold_train_02)):
    if scaffold_train_02[i] < c:
        avg_001.append(abs(scaffold_test_02[i]-scaffold_train_02[i]))
        break

for i in range(len(scaffold_train_05)):
    if scaffold_train_05[i] < c:
        avg_001.append(abs(scaffold_test_05[i]-scaffold_train_05[i]))
        break

for i in range(len(scaffold_train_08)):
    if scaffold_train_08[i] < c:
        avg_001.append(abs(scaffold_test_08[i]-scaffold_train_08[i]))
        break
for i in range(len(scaffold_train_pat)):
    if scaffold_train_pat[i] < c:
        avg_001.append(abs(scaffold_test_pat[i]-scaffold_train_pat[i]))
        break

c = 0.005
for i in range(len(scaffold_train_iid)):
    if scaffold_train_iid[i] < c:
        avg_0005.append(abs(scaffold_test_iid[i]-scaffold_train_iid[i]))
        break

for i in range(len(scaffold_train_02)):
    if scaffold_train_02[i] < c:
        avg_0005.append(abs(scaffold_test_02[i]-scaffold_train_02[i]))
        break

for i in range(len(scaffold_train_05)):
    if scaffold_train_05[i] < c:
        avg_0005.append(abs(scaffold_test_05[i]-scaffold_train_05[i]))
        break

for i in range(len(scaffold_train_08)):
    if scaffold_train_08[i] < c:
        avg_0005.append(abs(scaffold_test_08[i]-scaffold_train_08[i]))
        break
for i in range(len(scaffold_train_pat)):
    if scaffold_train_pat[i] < c:
        avg_0005.append(abs(scaffold_test_pat[i]-scaffold_train_pat[i]))
        break
```

```python
plt.plot(ls1, avg_02, 'c', marker=(5, 2))
plt.plot(ls1, avg_008, 'b', marker=(5, 2))
plt.plot(ls1, avg_005, 'r', marker=(5, 2))
plt.plot(ls1[:4], avg_001, 'g', marker=(5, 2))
plt.plot(ls1[:4], avg_0005, 'y', marker=(5, 2))
plt.legend(['train loss 0.2', 'train loss 0.1', 'train loss 0.05', 'train loss 0.01', 'train loss 0.005'])
plt.xlabel(r'$\rho$, ' + 'heterogeneity level of local data distributions')
plt.ylabel('abs(test loss - train loss)')
plt.rcParams.update({'font.size': 14})
plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:,.2f}'))
plt.locator_params(axis='y', nbins=8)
plt.savefig('scaffold.pdf')
```
